# 109 - K Means Clustering with Python

**Unsupervised learning algorithm**

* Tries to classify data based on their similarity

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
```

```python
from sklearn.datasets import make_blobs
```

```python
# Generate random data that are clustered
data = make_blobs(
    n_samples=200,
    n_features=2,
    centers=4,
    cluster_std=1.8,
    random_state=101
)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

# Creating a K Means Clustering Algorithm

```python
from sklearn.cluster import KMeans
```

```python
# n_clusters - Expected number of clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
# Centers of the clusters
kmeans.cluster_centers_
```

```python
# Labels it believes to be true for the clusters
kmeans.labels_
```

```python
# Real Labels
data[1]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

## Comparison with K Means = 2

```python
# n_clusters - Expected number of clusters
kmeans_2 = KMeans(n_clusters=2)
kmeans_2.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans_2.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
