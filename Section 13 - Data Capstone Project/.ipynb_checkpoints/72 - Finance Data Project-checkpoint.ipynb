We'll focus on bank stocks and see how they progressed throughout the [financial crisis](https://en.wikipedia.org/wiki/Financial_crisis_of_2007%E2%80%9308) all the way to early 2016.

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
# Get data from 1st Jan 2006 to 1st Jan 2016
start_date = datetime.date(2006,1,1)
end_date = datetime.date(2016,1,1)
```

## Data

We need to get data using pandas datareader. We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo


```python
# Getting bank DataFrames
BAC = data.DataReader("BAC","yahoo",start_date,end_date)
C = data.DataReader("C","yahoo",start_date,end_date)
GS = data.DataReader("GS","yahoo",start_date,end_date)
JPM = data.DataReader("JPM",'yahoo',start_date,end_date)
MS = data.DataReader("MS",'yahoo',start_date,end_date)
WFC = data.DataReader("WFC",'yahoo',start_date,end_date)
```

```python
# Sorting Tickers
tickers = 'BAC C GS JPM MS WFC'.split()
tickers.sort()
```

```python
# Consolidating into one DataFrame
# bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
# Above method depreciated due to API inconsistencies

bank_stocks = pd.read_pickle('data/bank_data')

# Setting Column Name
bank_stocks.columns.names = ['Bank Ticker','Stock Info']

bank_stocks.head()
```

# EDA

Let's explore the data a bit! Before continuing, I encourage you to check out the documentation on [Multi-Level Indexing](http://pandas.pydata.org/pandas-docs/stable/advanced.html) and [Using .xs](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.xs.html).
Reference the solutions if you can not figure out how to use .xs(), since that will be a major part of this project.

**What is the max Close price for each bank's stock throughout the time period?**

```python
close_df = bank_stocks.xs(key='Close',level=1,axis=1)
max_close_df = close_df.max()
print(max_close_df)
```

**Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
# Creating a new DataFrame called 'returns'
returns = pd.DataFrame()

for ticker in tickers:
    returns[f'{ticker} Return'] = close_df[ticker].pct_change()
```

**Create a pairplot using seaborn of the returns dataframe. What stock stands out to you? Can you figure out why?**

```python
import seaborn as sns
sns.set_style('whitegrid')
sns.pairplot(returns)
```

**Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. You should notice that 4 of the banks share the same day for the worst drop, did anything significant happen that day?**

```python
# Best Returns
max_return_date = returns.idxmax()
max_return_date
```

```python
# Worst Returns
min_return_date = returns.idxmin()
min_return_date
```

**You should have noticed that Citigroup's largest drop and biggest gain were very close to one another, did anything significant happen in that time frame?**

A stock split.

**Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2015?**

```python
# Riskiest overall - Highest Standard Deviation
std_all = returns.std()
std_all

# Riskiest is Citibank
```

```python
# Riskiest in 2015
returns_2015 = returns.loc['2015-01-01':'2016-01-01']
returns_2015.std()
```

**Create a distplot using seaborn of the 2015 returns for Morgan Stanley**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.distplot(returns_2015['MS Return'],bins=100)
```

**Create a distplot using seaborn of the 2008 returns for CitiGroup**

```python
citi_2008 = returns.loc['2008-01-01':'2009-01-01']
#plt.figure(figsize=(12,6))
sns.distplot(citi_2008['C Return'])
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
# Filter DataFrame to 2008 data
BAC_2008 = BAC.loc['2008-01-01':'2009-01-01']

# Set Figure Size
plt.figure(figsize=(12,6))

# Plot Rolling Average
BAC_2008['Close'].rolling(window=30).mean().plot(label='30 Day Average')

# Plot BAC Closing Price
BAC_2008['Close'].plot(label='BAC Close')

# Show Legend
plt.legend()
```

**Create a heatmap of the correlation between the stocks Close Price.**

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=close_df.corr(),annot=True)
```
